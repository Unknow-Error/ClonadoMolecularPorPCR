```python
import random

def nombre_a_ADN(nombre):
  """Convierte un nombre en una secuencia de ADN.

  Args:
    nombre: El nombre que se va a convertir.

  Returns:
    Una cadena que representa la secuencia de ADN.
  """
  # Crear un diccionario para mapear cada letra a una secuencia de ADN
  mapa_ADN = {
      'A': 'GCT',
      'B': 'CGA',
      'C': 'TAG',
      'D': 'ATC',
      'E': 'ACT',
      'F': 'GCA',
      'G': 'TAC',
      'H': 'ATG',
      'I': 'AGT',
      'J': 'CGC',
      'K': 'TGA',
      'L': 'GTC',
      'M': 'CTG',
      'N': 'ACG',
      'O': 'GAT',
      'P': 'AGC',
      'Q': 'TCG',
      'R': 'TGC',
      'S': 'GGC',
      'T': 'CCA',
      'U': 'CAC',
      'V': 'GTA',
      'W': 'AGG',
      'X': 'CCG',
      'Y': 'TTC',
      'Z': 'CTC',
      ' ': 'AAA'  # Espacio
  }

  # Convertir el nombre a mayúsculas
  nombre = nombre.upper()

  # Convertir cada letra del nombre a su secuencia de ADN correspondiente
  secuencia_ADN = ''.join([mapa_ADN[letra] for letra in nombre])

  return secuencia_ADN

def generar_marca_agua_azarosa(longitud):
  """Genera una secuencia aleatoria de ADN de una longitud dada.

  Args:
    longitud: La longitud deseada de la secuencia de marca de agua.

  Returns:
    Una cadena que representa la secuencia de marca de agua.
  """
  bases = ['A', 'T', 'G', 'C']
  marca_agua = ''.join(random.choice(bases) for _ in range(longitud))
  return marca_agua

# Ejemplo de uso
marca_de_agua = "LABORATORIO LOS TRES MOSQUETEROS"
marca_de_agua_ADN = nombre_a_ADN(marca_de_agua)
print(f"Nombre: {marca_de_agua}")
print(f"Secuencia de ADN: {marca_de_agua_ADN}")

```

```python
#Programa que indica las condiciones PCR

import primer3 as mt
from Bio.Seq import Seq

class MaquinaPCR:
    #Va a dar las condiciones para la PCR dado unos primers y una secuencia a levantar
    def __init__(self, primer_fwd, primer_rev, primer_overlap_fwd, primer_overlap_rev, sequence, temperatura_desnaturalizacion = 95, temperatura_anneling = None, temperatura_extension = 72, num_ciclos_molde = 5, num_ciclos_soe = 25, tiempo_extension = None, tiempo_desnaturalizacion_inicial = 5, tiempo_extension_final = 10):
        self.num_ciclos_molde = num_ciclos_molde
        self.num_ciclos_soe = num_ciclos_soe
        self.temperatura_desnaturalizacion = temperatura_desnaturalizacion
        self.temperatura_anneling = temperatura_anneling
        self.temperatura_extension = temperatura_extension
        self.primer_fwd = primer_fwd
        self.primer_rev = primer_rev
        self.primer_overlap_fwd = primer_overlap_fwd
        self.primer_overlap_rev = primer_overlap_rev
        self.sequence = sequence
        self.tiempo_desnaturalizacion_inicial = tiempo_desnaturalizacion_inicial
        self.tiempo_extension_final = tiempo_extension_final
        self.tiempo_extension = tiempo_extension

        self.tm_primer_fwd = mt.calc_tm(primer_fwd)
        self.tm_primer_rev = mt.calc_tm(primer_rev)
        self.tm_primer_overlap_fwd = mt.calc_tm(primer_overlap_fwd)
        self.tm_primer_overlap_rev = mt.calc_tm(primer_overlap_rev)

    def simulador_PCR(self):
        #Simular la PCR
        print(f"Condiciones de termociclado para la SOE-PCR para la secuencia: {self.sequence} \n")
        print(f"1 - Desnaturalización Inicial: {self.temperatura_desnaturalizacion}°C durante {self.tiempo_desnaturalizacion_inicial} minitos. \n")

        print(f"2 - {self.num_ciclos_molde} Ciclos para el molde: \n")
        print(f" - Desnaturalización: 95°C por 30 segundos")
        print(f" - Anneling - molde : {max(self.tm_primer_fwd, self.tm_primer_rev):.2f}°C por 30 segundos")
        print(f" - Extension: {self.temperatura_extension}°C por {self.tiempo_extension} minutos")

        print(f"3 - {self.num_ciclos_soe} Ciclos para los productos: \n")
        print(f" - Desnaturalización: 95°C por 30 segundos")
        print(f" - Anneling - SOE : {min(self.tm_primer_overlap_fwd, self.tm_primer_overlap_rev):.2f}°C por 30 segundos")
        print(f" - Extension: {self.temperatura_extension}°C por {self.tiempo_extension} minutos")

        print(f"4 - Extensión final: {self.temperatura_extension}°C durante {self.tiempo_extension_final} minutos. ")

    def get_PCR_Producto(self):
        fragmento_inicial = self.primer_overlap_fwd[:-len(self.primer_fwd)]
        fragmento_final = self.primer_overlap_rev[len(self.primer_rev):]
        return fragmento_inicial + self.sequence + fragmento_final

#PCR Plásmido #1
tiempo_extension_seq = len(secuencia_PROM_ENHAM_CMV_Intro_beta_Globina)/1000

pcr = MaquinaPCR(primers_secuencia_ENHAM_CMV_Intro_beta_Globina[0].seq, str(Seq(primers_secuencia_ENHAM_CMV_Intro_beta_Globina[1].seq).reverse_complement()), primer_pCMV_fwd, str(Seq(primers_secuencia_ENHAM_CMV_Intro_beta_Globina[1].seq).reverse_complement())+primer_pCMV_rev[len(primers_secuencia_ENHAM_CMV_Intro_beta_Globina[1].seq):], secuencia_PROM_ENHAM_CMV_Intro_beta_Globina, tiempo_extension = tiempo_extension_seq)
pcr.simulador_PCR()
print(pcr.get_PCR_Producto())

tiempo_extension_seq_2 = len(secuencia_pT7_ATTR1_LAC_5_UTR)/1000
pcr_2 = MaquinaPCR(primers_secuencia_pT7_ATTR1_LAC_5_UTR[0].seq, str(Seq(primers_secuencia_pT7_ATTR1_LAC_5_UTR[0].seq).reverse_complement()), primer_attr1_fwd, str(Seq(primers_secuencia_pT7_ATTR1_LAC_5_UTR[1].seq).reverse_complement())+primers_attr1_rev[len(primers_secuencia_pT7_ATTR1_LAC_5_UTR[1].seq):], secuencia_pT7_ATTR1_LAC_5_UTR, tiempo_extension = tiempo_extension_seq_2)
pcr_2.simulador_PCR()
print(pcr_2.get_PCR_Producto())

tiempo_extension_seq_3 = len(secuencia_ORF_CCBD_ATTR2)/1000
pcr_3 = MaquinaPCR(primers_secuencia_ORF_CCBD_ATTR2[0].seq, str(Seq(primers_secuencia_ORF_CCBD_ATTR2[1].seq).reverse_complement()), primer_ccdB_fwd, str(Seq(primers_secuencia_ORF_CCBD_ATTR2[1].seq).reverse_complement())+primer_attR2_rev[len(primers_secuencia_ORF_CCBD_ATTR2[1].seq):], secuencia_ORF_CCBD_ATTR2, tiempo_extension = tiempo_extension_seq_3)
pcr_3.simulador_PCR()
print(pcr_3.get_PCR_Producto())

tiempo_extension_seq_4 = len(secuencia_IRES_2)/1000
pcr_4 = MaquinaPCR(primers_secuencia_IRES[0].seq, str(Seq(primers_secuencia_IRES[1].seq).reverse_complement()), primer_orf_C_IRES_fwd, str(Seq(primers_secuencia_IRES[1].seq).reverse_complement())+primer_orf_C_IRES_rev[len(primers_secuencia_IRES[1].seq):], secuencia_IRES_2, tiempo_extension = tiempo_extension_seq_4)
pcr_4.simulador_PCR()
print(pcr_4.get_PCR_Producto())

tiempo_extension_seq_5 = len(secuencia_ORF_C)/1000
pcr_5 = MaquinaPCR(primers_ORF_C[0].seq, str(Seq(primers_ORF_C[1].seq).reverse_complement()), primer_orf_C_fwd, str(Seq(primers_ORF_C[1].seq).reverse_complement())+primer_orf_C_rev[len(primers_ORF_C[1].seq):], secuencia_ORF_C, tiempo_extension = tiempo_extension_seq_5)
pcr_5.simulador_PCR()
print(pcr_5.get_PCR_Producto())

pcr_6 = MaquinaPCR(primers_secuencia_IRES[0].seq, str(Seq(primers_secuencia_IRES[1].seq).reverse_complement()), primer_orf_E3_IRES_fwd, str(Seq(primers_secuencia_IRES[1].seq).reverse_complement())+primer_orf_E3_IRES_rev[len(primers_secuencia_IRES[1].seq):], secuencia_IRES_2, tiempo_extension = tiempo_extension_seq_4)
pcr_6.simulador_PCR()
print(pcr_6.get_PCR_Producto())

tiempo_extension_seq_6 = len(secuencia_ORF_E3)/1000
pcr_7 = MaquinaPCR(primers_ORF_E3[0].seq, str(Seq(primers_ORF_E3[1].seq).reverse_complement()), primer_orf_E3_fwd, str(Seq(primers_ORF_E3[1].seq).reverse_complement())+primer_orf_E3_rev[len(primers_ORF_E3[1].seq):], secuencia_ORF_E3, tiempo_extension = tiempo_extension_seq_6)
pcr_7.simulador_PCR()
print(pcr_7.get_PCR_Producto())

pcr_8 = MaquinaPCR(primers_secuencia_IRES[0].seq,str(Seq(primers_secuencia_IRES[1].seq).reverse_complement()), primer_orf_6K_IRES_fwd, str(Seq(primers_secuencia_IRES[1].seq).reverse_complement())+primer_orf_6K_IRES_rev[len(primers_secuencia_IRES[1].seq):],secuencia_IRES_2, tiempo_extension = tiempo_extension_seq_4)
pcr_8.simulador_PCR()
print(pcr_8.get_PCR_Producto())

tiempo_extension_seq_7 = len(secuencia_ORF_PROT_6K)/1000
pcr_9 = MaquinaPCR(primers_ORF_PROT_6K[0].seq, str(Seq(primers_ORF_PROT_6K[1].seq).reverse_complement()), primer_orf_6K_fwd, str(Seq(primers_ORF_PROT_6K[1].seq).reverse_complement())+primer_orf_6K_rev[len(primers_ORF_PROT_6K[1].seq):], secuencia_ORF_PROT_6K, tiempo_extension = tiempo_extension_seq_7)
pcr_9.simulador_PCR()
print(pcr_9.get_PCR_Producto())

pcr_10 = MaquinaPCR(primers_secuencia_IRES[0].seq, str(Seq(primers_secuencia_IRES[1].seq).reverse_complement()), primer_orf_E1_IRES_fwd, str(Seq(primers_secuencia_IRES[1].seq).reverse_complement())+primer_orf_E1_IRES_rev[len(primers_secuencia_IRES[1].seq):], secuencia_IRES_2, tiempo_extension = tiempo_extension_seq_4)
pcr_10.simulador_PCR()
print(pcr_10.get_PCR_Producto())

tiempo_extension_seq_8 = len(secuencia_ORF_E1)/1000
pcr_11 = MaquinaPCR(primers_ORF_E1[0].seq, str(Seq(primers_ORF_E1[1].seq).reverse_complement()), primer_orf_E1_fwd, str(Seq(primers_ORF_E1[1].seq).reverse_complement())+primer_orf_E1_rev_bGH_poliA[len(primers_ORF_E1[1].seq):], secuencia_ORF_E1_bGH_PoliA, tiempo_extension = tiempo_extension_seq_8)
pcr_11.simulador_PCR()
print(pcr_11.get_PCR_Producto())

tiempo_extension_seq_9 = len(secuencia_pUC)/1000
pcr_12 = MaquinaPCR(primers_secuencia_pUC[0].seq, str(Seq(primers_secuencia_pUC[1].seq).reverse_complement()), primer_orf_pUC_fwd, str(Seq(primers_secuencia_pUC[1].seq).reverse_complement())+primer_orf_pUC_rev[len(primers_secuencia_pUC[1].seq):], secuencia_pUC, tiempo_extension = tiempo_extension_seq_9)
pcr_12.simulador_PCR()
print(pcr_12.get_PCR_Producto())

tiempo_extension_seq_10 = len(secuencia_promotor_PC_5_UTR)/1000
pcr_13 = MaquinaPCR(primers_secuencia_promotor_PC_5_UTR[0].seq, str(Seq(primers_secuencia_promotor_PC_5_UTR[1].seq).reverse_complement()), primer_pKan_fwd, str(Seq(primers_secuencia_promotor_PC_5_UTR[1].seq).reverse_complement())+primer_pKan_rev[len(primers_secuencia_promotor_PC_5_UTR[1].seq):], secuencia_promotor_PC_5_UTR, tiempo_extension = tiempo_extension_seq_10)
pcr_13.simulador_PCR()
print(pcr_13.get_PCR_Producto())

tiempo_extension_seq_11 = len(secuencia_KANA)/1000
pcr_14 = MaquinaPCR(primers_secuencia_KANA[0].seq, str(Seq(primers_secuencia_KANA[1].seq).reverse_complement()), primer_orf_Kan_fwd, str(Seq(primers_secuencia_KANA[1].seq).reverse_complement())+primer_orf_Kan_rev[len(primers_secuencia_KANA[1].seq):], secuencia_KANA, tiempo_extension = tiempo_extension_seq_11)
pcr_14.simulador_PCR()
print(pcr_14.get_PCR_Producto())

tiempo_extension_seq_12 = len(secuencia_3_UTR_tKAN)/1000
pcr_15 = MaquinaPCR(primers_secuencia_3_UTR_tKANA[0].seq, str(Seq(primers_secuencia_3_UTR_tKANA[1].seq).reverse_complement()), primer_tKan_fwd, str(Seq(primers_secuencia_3_UTR_tKANA[1].seq).reverse_complement())+primer_tKan_rev[len(primers_secuencia_3_UTR_tKANA[1].seq):], secuencia_3_UTR_tKAN, tiempo_extension = tiempo_extension_seq_12)
pcr_15.simulador_PCR()
print(pcr_15.get_PCR_Producto())

```

```python
#Simular sistema Lambda-RED : Fusionar todas las secuencas

def fusion_lambda_RED(seq1, seq2):
    """
    Fusiona dos secuencias por regiones homólogas, eliminando la repetición de la secuencia homóloga.

    :param seq1: Primera secuencia.
    :param seq2: Segunda secuencia.
    :return: Secuencia fusionada.
    """
    # Buscar la región homóloga
    # Suponemos que la región homóloga está al final de seq1 y al inicio de seq2
    max_overlap = 0
    for i in range(1, min(len(seq1), len(seq2)) + 1):
        if seq1[-i:] == seq2[:i]:
            max_overlap = i

    # Fusionar las secuencias eliminando la repetición de la región homóloga
    secuencia_fusion = seq1 + seq2[max_overlap:]

    return secuencia_fusion

#Red-Lambda Fusión Plásmido #1

plasmido_1 =  fusion_lambda_RED(pcr.get_PCR_Producto(), pcr_2.get_PCR_Producto())
lista_fusion_red = [pcr_3.get_PCR_Producto(), pcr_4.get_PCR_Producto(), pcr_5.get_PCR_Producto(), pcr_6.get_PCR_Producto(),
                    pcr_7.get_PCR_Producto(), pcr_8.get_PCR_Producto(), pcr_9.get_PCR_Producto(), pcr_10.get_PCR_Producto(),
                    pcr_11.get_PCR_Producto(), pcr_12.get_PCR_Producto(), marca_agua_overlap, pcr_13.get_PCR_Producto(), pcr_14.get_PCR_Producto(), pcr_15.get_PCR_Producto()]

for i in lista_fusion_red:
    plasmido_1 = fusion_lambda_RED(plasmido_1, i)

print("Plásmido #1 : Nivel 2 para VLPs - Destino Gateway: \n")
print(plasmido_1)
print(len(plasmido_1))

#Red-Lambda Fusión Plásmido #2

plasmido_2 = fusion_lambda_RED(pcr_16.get_PCR_Producto(),pcr_17.get_PCR_Producto())
lista_fusion_red_2 = [pcr_18.get_PCR_Producto(), pcr_19.get_PCR_Producto()]

for i in lista_fusion_red_2:
    plasmido_2 = fusion_lambda_RED(plasmido_2, i)

print("\nPlásmido #2 : Sistema de permanencia y segregación - Adición por loxP: \n")
print(plasmido_2)
print(len(plasmido_2))
```

```python
#Simulación de Recombinación Cre/loxP

def cre_loxP_recombinacion(sec1, sec2, secuencia_loxP):
    #Buscar los sitios loxP en las secuencias
    indice_inicio_loxP1 = sec1.find(secuencia_loxP)
    indice_fin_loxP1 = indice_inicio_loxP1 + len(secuencia_loxP)
    indice_inicio_loxP2 = sec2.find(str(Seq(secuencia_loxP).reverse_complement()))
    indice_fin_loxP2 = indice_inicio_loxP2 + len(secuencia_loxP)

    # Verificar la orientación de los sitios loxP
    orientacion_loxP1 = 1  # Asumiendo orientación forward por defecto
    orientacion_loxP2 = 1  # Asumiendo orientación forward por defecto

    # Verificar si el segundo sitio loxP está en orientación inversa
    if sec2[indice_inicio_loxP2:indice_fin_loxP2] != secuencia_loxP:
      orientacion_loxP2 = -1  # Orientación inversa

    # Verificar si se encontraron sitios loxP en ambas secuencias
    if indice_inicio_loxP1 != -1 and indice_inicio_loxP2 != -1:
      #Extraer las secuencias a recombinar
      secuencia_antes_loxP1 = sec1[:indice_inicio_loxP1]
      secuencia_despues_loxP1 = sec1[indice_fin_loxP1:]
      secuencia_antes_loxP2 = sec2[:indice_inicio_loxP2]
      secuencia_despues_loxP2 = sec2[indice_fin_loxP2:]

      # Realizar la recombinación
      if orientacion_loxP1 == orientacion_loxP2:  # Misma orientación: intercambio
        secuencia_recombinada = secuencia_antes_loxP1 + secuencia_loxP + secuencia_antes_loxP2 + secuencia_despues_loxP2 + secuencia_loxP + secuencia_despues_loxP1
      else:  # Orientación inversa: inversión e intercambio
        secuencia_recombinada = secuencia_antes_loxP1 + secuencia_loxP + str(Seq(secuencia_antes_loxP2).reverse_complement()) + str(Seq(secuencia_despues_loxP2).reverse_complement()) + secuencia_loxP + secuencia_despues_loxP1

      return secuencia_recombinada

    else:
      print("Secuencia loxP no encontrada en ambas secuencias")

plasmido_3_postloxP = cre_loxP_recombinacion(plasmido_1, plasmido_2, secuencia_loxP)
print(plasmido_3_postloxP)
print(len(plasmido_3_postloxP))
```
